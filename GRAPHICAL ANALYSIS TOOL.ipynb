## MATPLOTLIB

Matplotlib is a plotting library for the Python Programming Languages and its numerical mathematics extension NumPy.
It provides an object oriented API for embedding plots into applications .


Advantages of MatPlotLib Library are

1. Generally easy to get started for simple plots.
2. Support for custom labels and text.
3. Great control of every element in a figure.
4. High-Quality output in many formats.
5. Very customizable in general.

## Creating Subplots

it means that within one canvas i can create multi subplot

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'bo')
plt.subplot(2,2,3)
plt.plot(x,y,'go--')
plt.subplot(2,2,4)
plt.plot(x,y,'b*--')
```

```python
## Compute the x and y coordinates for points on a sine curve
```

```python
x=np.arange(0,4*np.pi,0.1)  ## taking the values from 0 to 4pi and the third parameter is the step size
y=np.sin(x)
plt.title('Sine wave form')

## plot the points using matplotlib

plt.plot(x,y)
plt.show()
```

```python
## Subplot

## Compute the x and y coordinates for points on a sine curve and cose curve

x=np.arange(0,5*np.pi,0.1)  ## taking the values from 0 to 4pi and the third parameter is the step size
y_sin=np.sin(x)
y_cos=np.cos(x)

## Set up the subplot grid that has height 2 and width 1
## and set the first subplot as active

plt.subplot(2,1,1)

# Make the first plot
plt.plot(x,y_sin,'r--')
plt.title('Sine wave')

## Set the second subplot as active and make the second plot

plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('Cos Wave')


## Show the figure

plt.show()
```

## Bar Plot

```python
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]
   
plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()

```

## Box Plot Using MatPlotLib

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

## Rectangular Box Plot

plt.boxplot(data,vert=True,patch_artist=True)
```

# PIE CHART

```python
## DATA TO PLOT

labels='Python','C++','Ruby','Java'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0) #explode 1st list

#Plot
plt.pie(sizes,explode=explode,labels=labels,colors=colors,
autopct='%1.1f%%',shadow=True)

plt.axis('equal')
plt.show()
```

## Correlation with Heatmap

A correlation heatmap uses colored cells , typically in a monochromatic scale,to show a 2D corrlation matrix(table) between
two discrete dimensions or event types. It is very important in feature scaling

Whenever we are trying to apply correlation the value will be ranging between -1 and 1 and that is called Spearson's
Coefficient Correlation

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

## Joint Plot

A Joint Plot Allows to study the relationship between 2 numeric variables. The central chart display their correlation . It 
usually a scatter plot, a hexbin plot, a 2D histogram or a 2D density plot.

```python
numeric_feature=[feature for feature in df1.columns if df1[feature].dtypes != 'O']
numeric_feature
```

```python
plt.figure(figsize=(8,8))
plt.suptitle('Univariate Analysis on Numeric Feature',fontsize=20,fontweight='bold')

for i in range(len(numeric_feature)):
    plt.subplot(2,2,i+1)
    sns.kdeplot(x=df1[numeric_feature[i]],shade=True,color='b')
    plt.xlabel(numeric_feature[i])
```
